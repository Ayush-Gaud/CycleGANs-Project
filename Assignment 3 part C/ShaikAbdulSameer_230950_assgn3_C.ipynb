```python
def split_indices(n,f,s):
    n_test=int(n*f)
    np.random.seed(s)
    idxs=np.random.permutation(n)
    return idxs[:n_test],idxs[n_test:]
```

```python
def encod(y,df):
    y_encod=[]
    for i in range(len(y)):
        if(y[i]==df['Team2'].iloc[i]):
            y_encod.append([0,1])
        else :
            y_encod.append([1,0])
    return np.array(y_encod)

y_train=encod(y_train,df)
y_test=encod(y_test,df)

print(y)
```
